```python
import cv2
import dlib
import os
```

```python
face_detector = dlib.get_frontal_face_detector()

directory = "Faces/"
f_directory = "processedFaces/"
s_dir = os.listdir(directory)
```

```python
def facecrop(image):

    img = cv2.imread(image)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    try:

        faces = face_detector(gray)

        for f in faces:

            x1 = f.left()
            y1 = f.top()
            x2 = f.right()
            y2 = f.bottom()

            sub_face = img[y1-50:y2+50, x1-50:x2+50]

            sub_face = cv2.resize(sub_face, (80,80))

            f_name = image.split('/')
            f_name = f_name[-1]

            global k
            global f_directory

            f_directory = (f_directory + s_dir[k] + '/')
            cv2.imwrite(f_directory + f_name, sub_face)
            print ("Writing: " + image)
            f_directory = "processedFaces/"

    except:
        print("error")
        pass
```

```python
if __name__ == '__main__':
    for k in range(len(s_dir)):
        directory = (directory+s_dir[k]+'/')
        images = os.listdir(directory)
        i = 0
        
        for img in images:
            file = directory + img
            print(file)
            print (i)
            facecrop(file)
            i += 1

        directory = "Faces/"
        f_directory = "processedFaces/"
```
